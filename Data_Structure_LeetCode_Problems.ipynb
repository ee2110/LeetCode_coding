# **Max Consecutive Ones**

Given a binary array nums, return the maximum number of consecutive 1's in the array.

```python
# Example

'''
Input: nums = [1,1,0,1,1,1]
Output: 3
Explanation: The first two digits or the last three digits are consecutive 1s. 
The maximum number of consecutive 1s is 3.

Constraints:
1 <= nums.length <= 105
nums[i] is either 0 or 1.
'''
```

```python
'''
strategy
1. iterate all element in array
2. check if the current num is 1, if yes, count plus 1.
3. check if num is 1 or 0
4. If is 1, count + 1; if is 0, count = 0
5. check the total consecutive count one, if is greater than previous, replace the max count value 
'''

def findMaxConsecutiveOnes(nums):
    max_cons = 0
    one_count = 0

    for i in nums:
      if i == 1:
        one_count+=1

      if i == 0:
        one_count = 0

      if max_cons < one_count:
        max_cons = one_count

    return(max_cons)
```

```python
nums = [1,1,0,1,1,1,1,1,0,0,1,1,1,1,0,0,1,1,1]
```

```python
print('The maximum number of consecutive 1s is: ', findMaxConsecutiveOnes(nums))
```

# **Squares of a Sorted Array**

Given an integer array nums sorted in non-decreasing order, return an array of the squares of each number sorted in non-decreasing order.

```python
# Example

'''
Input: nums = [-4,-1,0,3,10]
Output: [0,1,9,16,100]
Explanation: After squaring, the array becomes [16,1,0,9,100].
After sorting, it becomes [0,1,9,16,100].
'''
```

```python
nums = [-7,-3,2,3,11]
```

```python
def sortedSquares_1(nums):

  squared_num = []
  for i in nums:
    squared_num.append(i*i)

  sorted_num = []
  copy_squared_num = squared_num

  while squared_num:
      min_num = copy_squared_num[0]
      
      for x in copy_squared_num: 
          if x < min_num:
            min_num = x

      sorted_num.append(min_num)
      copy_squared_num.remove(min_num)

  return sorted_num
```

```python
sortedSquares_1(nums)
```

```python
input_arr = [1,0,2,3,0,4,5,0]
zero_dup = 0
length_ = len(input_arr)-1

for l in range(len(input_arr)):

  # Stop when left points beyond the last element in the original list
  # which would be part of the modified list
  if l > length_ - zero_dup: #number of elements which would be included in the final array.
      break

  # Count the zeros
  if input_arr[l] == 0:

      # Edge case: This zero can't be duplicated. We have no more space,
      # as left is pointing to the last element which could be included 
      if l == length_ - zero_dup: 
          input_arr[length_] = 0
          length_ -= 1
          break

      zero_dup += 1

print('after first checking, the zero duplicate is ', zero_dup)

# Start backwards from the last element which would be part of new list.
last = length_ - zero_dup
print('The last element index is ', last)

print('The original array is ', input_arr)
# Copy zero twice, and non zero once.
for i in range(last, -1, -1):
    print('\n')
    print('Now element index is ', i)

    if input_arr[i] == 0:
        print('Found zero at index ', i)
        print('Target element be replace with zero is at index ', i + zero_dup)
        input_arr[i + zero_dup] = 0
        print(input_arr)

        zero_dup -= 1
        print('The zero duplicate left :',zero_dup)

        print('Target element be replace with zero is at index ', i + zero_dup)
        input_arr[i + zero_dup] = 0
        print(input_arr)

    else:
        print('Found non-zero at index ', i)
        print('Target element is at index ', i + zero_dup)
        print('Replace the element of {} with {}' .format(input_arr[i + zero_dup],input_arr[i]))
        input_arr[i + zero_dup] = input_arr[i]
        print(input_arr)
  
```

# **Merge Sorted Array**

You are given two integer arrays nums1 and nums2, sorted in non-decreasing order, and two integers m and n, representing the number of elements in nums1 and nums2 respectively.

Merge nums1 and nums2 into a single array sorted in non-decreasing order.

The final sorted array should not be returned by the function, but instead be stored inside the array nums1. To accommodate this, nums1 has a length of m + n, where the first m elements denote the elements that should be merged, and the last n elements are set to 0 and should be ignored. nums2 has a length of n.

```python
# Example 1

'''
Input: nums1 = [1,2,3,0,0,0], m = 3, nums2 = [2,5,6], n = 3
Output: [1,2,2,3,5,6]
Explanation: The arrays we are merging are [1,2,3] and [2,5,6].
The result of the merge is [1,2,2,3,5,6] with the underlined elements coming from nums1.
'''

# Example 2
'''
Input: nums1 = [1], m = 1, nums2 = [], n = 0
Output: [1]
Explanation: The arrays we are merging are [1] and [].
The result of the merge is [1].
'''

#Example 3
'''
Input: nums1 = [0], m = 0, nums2 = [1], n = 1
Output: [1]
Explanation: The arrays we are merging are [] and [1].
The result of the merge is [1].
Note that because m = 0, there are no elements in nums1. The 0 is only there to ensure the merge result can fit in nums1.
'''
```

```python
def generate_sample_input(x):
  if x == 1:
    nums1 = [1,2,3,0,0,0]
    m = 3
    nums2 = [2,5,6]
    n = 3

  if x == 2:
    nums1 = [1]
    m = 1 
    nums2 = []
    n = 0

  if x == 3:
    nums1 = [0]
    m = 0
    nums2 = [1]
    n = 1

  if x == 4:
    nums1 = [4,0,0,0,0,0]
    m = 1
    nums2 = [1,2,3,5,6]
    n = 5

  return nums1, m, nums2, n
```

```python
def merge(nums1, m, nums2, n):
    """
    :type nums1: List[int]
    :type m: int
    :type nums2: List[int]
    :type n: int
    :rtype: None Do not return anything, modify nums1 in-place instead.
    """

    if m > 0 and n > 0:
      del nums1[m:]
        
    elif m == 0 and n > 0:
      del nums1[:]

    for j in range(n):
      nums1.insert(m+j, nums2[j])
      
    nums1.sort()
```

```python
nums1, m, nums2, n = generate_sample_input(4)
merge(nums1, m, nums2, n)
nums1
```
